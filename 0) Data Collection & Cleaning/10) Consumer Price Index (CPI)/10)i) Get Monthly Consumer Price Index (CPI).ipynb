```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import pprint as pp
import requests
```

```python
# Get consumer price index (CPI)-  Monthly

api_key = 'YOUR_API_KEY'

url = f'https://www.alphavantage.co/query?function=CPI&interval=monthly&apikey={api_key}'
r = requests.get(url)
data = r.json()
```

```python
monthly_consumer_price_index_df = pd.DataFrame(data['data'])
monthly_consumer_price_index_df.head()
```
