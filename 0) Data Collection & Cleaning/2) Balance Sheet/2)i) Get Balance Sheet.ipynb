```python
import pandas as pd
import requests
import pprint as pp
```

```python
# Get Balance Sheet - Quarterly & Annual

api_key = 'YOUR_API_KEY'
ticker = 'AMZN'
url = f'https://www.alphavantage.co/query?function=BALANCE_SHEET&symbol={ticker}&apikey={api_key}'
r = requests.get(url)
data = r.json()
```
