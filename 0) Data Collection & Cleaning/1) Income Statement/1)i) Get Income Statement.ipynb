```python
import pandas as pd
import requests
import pprint as pp
```

```python
# Get Income Statement - Quarterly & Annual

api_key = 'YOUR_API_KEY'
ticker = 'AMZN'
url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={api_key}'
response = requests.get(url)
data = response.json()
```
